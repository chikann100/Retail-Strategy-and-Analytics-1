```python
## Checking for unique values and their counts in the 'LIFESTAGE' column
lifestage_counts = df['LIFESTAGE'].value_counts()
```

```python
#plotting the prportion of customers based on life stage
plt.figure(figsize=(10, 6))
lifestage_counts.plot(kind='bar', color='skyblue')
plt.title('Proportion of Customers Based on Life Stage')
plt.xlabel('Life Stage')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()
```

```python
# showing distribution of customers across different life stages and premium customer categories

pivot_table = pd.pivot_table(df, index='LIFESTAGE', columns='PREMIUM_CUSTOMER', 
                             aggfunc='count', fill_value=0)['LYLTY_CARD_NBR']

# Plotting the bar chart
pivot_table.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Customer Distribution by Life Stage and Premium Category')
plt.xlabel('Life Stage')
plt.ylabel('Number of Customers')
plt.legend(title='Premium Category')

# Show the plot
plt.show()
```

```python
df['DATE'] = pd.to_datetime(df['DATE'], format='%d-%m-%y')
```

```python
sales_over_time = df.groupby('DATE')['TOT_SALES'].sum()
```

```python
plt.figure(figsize=(10, 6))
sales_over_time.plot()
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
sales_by_product = df.groupby('PREMIUM_CUSTOMER')['TOT_SALES'].sum()
total_sales_sum = df['TOT_SALES'].sum()

sales_proportion = (sales_by_product / total_sales_sum) * 100  # Multiply by 100 to convert to percentage

```

```python
plt.figure(figsize=(10, 6))
sales_proportion.sort_values(ascending=False).plot(kind='bar')
plt.title('Proportion of Sales ')
plt.xlabel('PREMIUM_CUSTOMER')
plt.ylabel('LIFESTAGE')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
